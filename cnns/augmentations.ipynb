# Image Augmentations

In order for a model to learn a diverse and generalizable representation of the data, we can augment our data to artificially increase our dataset size. Some examples of image augmentations are flips and color jitter. There are many more augmentations available but these are the most common.  Most models also like taking images of a set size by cropping an input image. Lets see some examples of augmentations.

```python
from PIL import Image
from torchvision import transforms
from IPython.display import display

make_tensor = transforms.ToTensor()
make_pil = transforms.ToPILImage()

v_flip = transforms.RandomVerticalFlip()
h_flip = transforms.RandomHorizontalFlip()

jitter = transforms.ColorJitter(30, 30, 30, (-0.5, 0.5)) # these are extreme values

crop = transforms.CenterCrop((300, 400))

img = Image.open("../images/perceptron.png").convert("RGB")
img = make_tensor(img)
img = v_flip(img)
img = h_flip(img)
img = jitter(img)
img = crop(img)

augmented = make_pil(img)
display(augmented)
```
